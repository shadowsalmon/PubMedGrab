```python
from Bio import Entrez

def search(query):
    Entrez.email = 'echoadvisor13@gmail.com'
    handle = Entrez.esearch(db = 'pubmed', sort = 'pub+date', retmax = '20',retmode = 'xml',term = query,
                           datetype = 'pdat', mindate = '1997')
    results = Entrez.read(handle)
    return results

def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'echoadvisor13@gmail.com'
    handle = Entrez.efetch(db = 'pubmed', retmode = 'xml', id = ids)
    results = Entrez.read(handle)
    return results
```
